**Elabore mais 6 questões e as responda com os dados.**

13. Qual foi o produto teve mais pedidos?

```python
df_vendas.groupby('Produto')['Quantidade'].sum()
```

14. Qual foi o faturamento de cada produto?

```python
df_vendas.groupby('Produto')['Valor Final'].sum().sort_values(ascending=False)
```

15. Qual loja teve o maior faturamento?

```python
df_vendas.groupby('ID Loja')['Valor Final'].sum().sort_values(ascending=False)
```

16. Qual a data teve mais vendas?

```python
df_vendas.groupby('Data')['Valor Final'].sum().sort_values(ascending=False)
```

17. Qual é o valor mais alto na coluna Valor Final?

```python
df_vendas['Valor Final'].max()
```

18. Existe alguma loja que vende mais produtos caros que as outras?

```python
df_vendas.groupby('ID Loja')['Valor Unitário'].mean().sort_values(ascending=False)
```
